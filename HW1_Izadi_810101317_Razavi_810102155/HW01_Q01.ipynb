```python
class Neuron:
  def __init__(self, w1, w2, tetha):
    self._w1 = w1
    self._w2 = w2
    self._tetha = tetha
    self._is_connected = False

  def Connect(self, n1, n2):
    self._n1 = n1
    self._n2 = n2
    self._is_connected = True;
    return self

  def AsFix(self, x1, x2):
    self._x1 = x1
    self._x2 = x2
    return self

  def Sum(self):
    return (self._x1 * self._w1) + (self._x2 * self._w2)

  def Actuate(self, sum):
    return 1 if ( sum >= self._tetha) else 0

  def Run(self):
    if self._is_connected:
      self._x1 = self._n1.Run()
      self._x2 = self._n2.Run()

    sum = self.Sum()

    return self.Actuate(sum)
```

```python
# these functions return logical gates.
def get_AND(n1, n2):
  ni_1 = Neuron(1, -1, 1)
  ni_2 = Neuron(-1, 1, 1)
  no = Neuron(1, 1, 1)

  ni_1.Connect(n1, n2)
  ni_2.Connect(n1, n2)
  no.Connect(ni_1, ni_2)

  return no

def get_OR(n1, n2):
  return Neuron(1, 1, 1).Connect(n1, n2)

#Identity Neurons (Run() = x)
def get_ID(x):
 return Neuron(1, 0, 1).AsFix(x, 0)
```
