```python
import torch
from torch import nn, optim


class VAE(nn.Module):
    def __init__(self):
        super(VAE, self).__init__()
        self.encoder = nn.Sequential(
            nn.Conv2d(3, 16, 3, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(16, 32, 3, stride=2, padding=1),
            nn.ReLU(),
            nn.Flatten(),
            nn.Linear(32 * 16 * 16, 128),
            nn.ReLU(),
            nn.Linear(128, 20)
        )
        self.decoder = nn.Sequential(
            nn.Linear(10, 128),
            nn.ReLU(),
            nn.Linear(128, 32 * 16 * 16),
            nn.ReLU(),
            nn.Unflatten(1, (32, 16, 16)),
            nn.ConvTranspose2d(32, 16, 3, stride=2, padding=1, output_padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(16, 3, 3, stride=2, padding=1, output_padding=1),
            nn.Sigmoid()
        )

    def reparameterize(self, mu, log_var):
        std = torch.exp(0.5 * log_var)
        eps = torch.randn_like(std)
        return mu + eps * std

    def forward(self, x):
        x = self.encoder(x)
        mu = x[:, :10]
        log_var = x[:, 10:]
        z = self.reparameterize(mu, log_var)
        return self.decoder(z), mu, log_var

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = VAE().to(device)
optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.MSELoss()


for epoch in range(20):
    for images, _ in dataloader:
        images = images.to(device)
        reconstruction, mu, log_var = model(images)
        loss = criterion(reconstruction, images) + \
               -0.5 * torch.sum(1 + log_var - mu.pow(2) - log_var.exp())
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    print(f'Epoch {epoch+1}, Loss: {loss.item()}')
```

```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, Dataset
import os
import random
from PIL import Image

class MultiFolderDataset(Dataset):
    def __init__(self, folder_paths, transform=None):
        self.folder_paths = folder_paths
        self.transform = transform
        self.image_paths = []
        self.labels = []
        self._load_images()

    def _load_images(self):
        for label, folder_path in enumerate(self.folder_paths):
            for root, _, files in os.walk(folder_path):
                for file in files:
                    if file.endswith(('jpg', 'jpeg', 'png')):
                        self.image_paths.append(os.path.join(root, file))
                        self.labels.append(label)

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image_path = self.image_paths[idx]
        image = Image.open(image_path).convert('RGB')
        label = self.labels[idx]

        if self.transform:
            image = self.transform(image)

        return image, label

root_dir = '/content/drive/MyDrive/NNDL_HW6/data/CartoonFaces/cartoonset100k_jpg'

selected_folders = random.sample(range(10), 5)
print("Selected folders:", selected_folders)

folder_paths = [os.path.join(root_dir, str(folder)) for folder in selected_folders]

transform = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.ToTensor(),
])


dataset = MultiFolderDataset(folder_paths, transform=transform)


dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
```

# Conditional VAE

```python
import torch
from torch import nn, optim

class CVAE(nn.Module):
    def __init__(self, num_conditions):
        super(CVAE, self).__init__()
        # Expanded input for encoder to include condition
        self.encoder = nn.Sequential(
            nn.Conv2d(3 + num_conditions, 16, 3, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(16, 32, 3, stride=2, padding=1),
            nn.ReLU(),
            nn.Flatten(),
            nn.Linear(32 * 16 * 16, 128),
            nn.ReLU(),
            nn.Linear(128, 20)  # Output mu and log_var
        )

        # Decoder takes both z and condition
        self.decoder = nn.Sequential(
            nn.Linear(10 + num_conditions, 128),
            nn.ReLU(),
            nn.Linear(128, 32 * 16 * 16),
            nn.ReLU(),
            nn.Unflatten(1, (32, 16, 16)),
            nn.ConvTranspose2d(32, 16, 3, stride=2, padding=1, output_padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(16, 3, 3, stride=2, padding=1, output_padding=1),
            nn.Sigmoid()
        )

    def reparameterize(self, mu, log_var):
        std = torch.exp(0.5 * log_var)
        eps = torch.randn_like(std)
        return mu + eps * std

    def forward(self, x, c):
        # Reshape and expand condition to match image dimensions
        batch_size, _, height, width = x.size()
        c = c.reshape(batch_size, -1, 1, 1)  # Reshape to (batch_size, num_conditions, 1, 1)
        c = c.expand(batch_size, -1, height, width)  # Expand to match x dimensions

        # Concatenate along the channel dimension
        x = torch.cat([x, c], dim=1)
        x = self.encoder(x)
        mu = x[:, :10]
        log_var = x[:, 10:]
        z = self.reparameterize(mu, log_var)

        # Concatenate z with the condition vector for the decoder
        z = torch.cat([z, c.reshape(batch_size, -1)], dim=1)
        return self.decoder(z), mu, log_var

# Assuming num_conditions is the size of the conditioning vector
num_conditions = 10  # Change this based on your dataset
model = CVAE(num_conditions=num_conditions)
optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.MSELoss()

# Example use in training
for epoch in range(3):
    for images, labels in dataloader:
        # Assuming labels are available and one-hot encoded
        c = torch.nn.functional.one_hot(labels, num_classes=num_conditions).float()

        # Forward pass with images and condition vector
        reconstruction, mu, log_var = model(images, c)

        # Compute the loss
        reconstruction_loss = criterion(reconstruction, images)
        kl_divergence = -0.5 * torch.sum(1 + log_var - mu.pow(2) - log_var.exp())
        loss = reconstruction_loss + kl_divergence

        # Backpropagation and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    print(f'Epoch {epoch+1}, Loss: {loss.item()}')
```
